```python
import cv2
import numpy as np
import os

def process_images(directory_path, roi_coordinates, roi_dimensions):
    # Get the list of image files in the directory
    image_files = [f for f in os.listdir(directory_path) if f.endswith(('.jpg', '.jpeg', '.png'))]

    # Process each image in the directory
    for image_file in image_files:
        # Load the image
        image_path = os.path.join(directory_path, image_file)
        image = cv2.imread(image_path)

        # Define the height and width of the original image
        original_height, original_width, _ = image.shape

        # Unpack ROI coordinates and dimensions
        top_left_x, top_left_y = roi_coordinates
        height, width = roi_dimensions

        # Extract the ROI coordinates based on the provided top-left coordinates
        x = top_left_x
        y = top_left_y

        # Create a mask for the selected ROI
        mask = np.zeros((original_height, original_width), dtype=np.uint8)
        mask[y:y+height, x:x+width] = 255

        # Convert the mask to uint8
        mask = mask.astype(np.uint8)

        # Apply the mask to the image
        masked_image = cv2.bitwise_and(image, image, mask=mask)

        # Create a directory for masked images if it doesn't exist
        masked_directory = os.path.join(directory_path, 'masked')
        if not os.path.exists(masked_directory):
            os.makedirs(masked_directory)

        # Generate the masked image filename
        masked_filename = os.path.join(masked_directory, os.path.splitext(image_file)[0] +  os.path.splitext(image_file)[1])

        # Save the masked image
        cv2.imwrite(masked_filename, masked_image)

# Parameters usage:
directory_path = r"C:\Users\kwayi\Desktop\School\MasterArbeite_23.24\Codes\Glenn__upssss\Glenn\16.05.2024_images\undistorted\right image"
#Adjust the ROI to fit area
roi_coordinates = (2398, 1567)
roi_dimensions = (3000, 2021)

process_images(directory_path, roi_coordinates, roi_dimensions)
```

```python
import cv2
import os

def crop_and_save_images(source_directory, destination_directory, top_left, bottom_right):
    # Check if destination directory exists, if not, create it
    if not os.path.exists(destination_directory):
        os.makedirs(destination_directory)

    # Iterate over all files in the source directory
    for filename in os.listdir(source_directory):
        if filename.lower().endswith((".png", ".jpg", ".jpeg")):  # Add or remove file types as needed
            # Construct the full file path
            file_path = os.path.join(source_directory, filename)
            # Read the image
            image = cv2.imread(file_path)
            # Check if the image was loaded successfully
            if image is None:
                print(f"Warning: Failed to load image {filename}. Skipping...")
                continue

            # Check if crop coordinates are within the image dimensions
            if top_left[0] >= bottom_right[0] or top_left[1] >= bottom_right[1]:
                print(f"Invalid crop coordinates for {filename}. Skipping...")
                continue

            height, width = image.shape[:2]
            if (top_left[0] < 0 or top_left[1] < 0 or
                bottom_right[0] > width or bottom_right[1] > height):
                print(f"Crop coordinates out of bounds for {filename}. Skipping...")
                continue

            # Crop the image
            cropped_image = image[top_left[1]:bottom_right[1], top_left[0]:bottom_right[0]]

            # Construct the destination file path
            dest_file_path = os.path.join(destination_directory, filename)
            # Save the cropped image
            cv2.imwrite(dest_file_path, cropped_image)
            print(f"Cropped image saved to {dest_file_path}")

if __name__ == "__main__":
    # Define the source directory
    source_dir = r"C:\Users\kwayi\Desktop\School\MasterArbeite_23.24\Codes\Glenn__upssss\Glenn\Lab_test_data\images\images\undistorted"
    # Define the destination directory
    dest_dir = r"C:\Users\kwayi\Desktop\School\MasterArbeite_23.24\Codes\Glenn__upssss\Glenn\Lab_test_data\images\images\undistorted\rigtht_calibrated_filtered_cropped"

    # Correct the order of the top left and bottom right coordinates
    # The top_left should have smaller x and y values than bottom_right
    top_left = (2654, 1060)  # Corrected values
    bottom_right = (3467, 1753)  # Corrected values

    # Crop and save the images
    crop_and_save_images(source_dir, dest_dir, top_left, bottom_right)
```
